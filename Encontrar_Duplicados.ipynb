```python
evento_duplicados = "/content/abononoligado.xlsx" #Evento del cual quieres saber qué lugares se duplicaron
evento_comparar = "/content/eventoconduplicidad.xlsx" #Evento que deseas usar para comparar con el evento que presenta duplicados
```

## Ejecuta el código

Una vez que ya hiciste los pasos anteriores, solo restará que corras el código para que genere la identificación de los duplicados, esto se hace de la siguiente manera, En la parte superior de la pantalla dando click en "Runtime" y luego en "Run all"

## Ignora esto
Esto es el código que realiza la inspección de duplicados, puedes ignorarlo y brincar al siguiente apartado abajo

```python
!pip install pandas
import pandas as pd

# Dataframe Gen
df1 = pd.read_excel(evento_duplicados)
df2 = pd.read_excel(evento_comparar)

# Gen column to add the duplicated status
df1['Duplicado'] = df1.apply(lambda row: 'duplicado' if any(df2[['ZONA', 'SECCIÓN', 'ASIENTO']].eq(row[['ZONA', 'SECCIÓN', 'ASIENTO']]).all(axis=1)) else '', axis=1)

# Show duplicated rows
duplicados = df1[df1['Duplicado'] == 'duplicado']
print("Duplicados encontrados:")
print(duplicados)

# Save the duplicated values in the first file
df1.to_excel(evento_duplicados, index=False)
```
